```python
def writehistory(text):
    with open(logfile, 'a', encoding='utf-8') as f:
        f.write(text)
        f.write('\n')
    f.close()
```

```python
def img2txt(input_text, input_image):

    # load the image
    image = Image.open(input_image)

    writehistory(f"Input text: {input_text} - Type: {type(input_text)} - Dir: {dir(input_text)}")
    if type(input_text) == tuple:
        prompt_instructions = """
        Describe the image using as much detail as possible, is it a painting, a photograph, what colors are predominant, what is the image about?
        Describe the image using as much detail as possible.
        You are a helpful AI assistant who is able to answer questions about the image.
        What is the image all about?
        If there is some person and asking who it is, find the name of the person.
        If the image has a specific name, answer the specific name and describe it
        """
    else:
        prompt_instructions = """
        Act as an expert in imagery descriptive analysis, using as much detail as possible from the image, respond to the following prompt:

        """ + input_text

    writehistory(f"prompt_instructions: {prompt_instructions}")
    prompt = "USER: <image>\n" + prompt_instructions + "\nASSISTANT:"

    outputs = pipe(image, prompt=prompt, generate_kwargs={"max_new_tokens": 200})
# Properly extract the response text
    if outputs is not None and len(outputs[0]["generated_text"]) > 0:
        match = re.search(r'ASSISTANT:\s*(.*)', outputs[0]["generated_text"])
        if match:
            # Extract the text after "ASSISTANT:"
            reply = match.group(1)
        else:
            reply = "No response found."
    else:
        reply = "No response generated."

    return reply
```

```python
from langdetect import detect
from gtts import gTTS
from gtts.lang import tts_langs
import whisper
import torch
import numpy as np

# Dynamically fetch all supported languages from gTTS
SUPPORTED_LANGUAGES = tts_langs()  # Returns a dictionary of all supported languages and their codes

# Automatically detect device
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Load the Whisper model
model = whisper.load_model("medium", device=DEVICE)

# Ensure full precision to avoid potential precision conflicts
if DEVICE == "cuda":
    model.float()  # Convert model weights to full precision (float32)

# Enhanced transcribe function
def transcribe(audio):
    # Check if the audio input is None or empty
    if audio is None or audio == '':
        return '', 'en'  # Default to English

    # Load and preprocess the audio
    audio = whisper.load_audio(audio)
    audio = whisper.pad_or_trim(audio)

    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # Detect the language of the spoken audio using Whisper
    _, probs = model.detect_language(mel)
    detected_language = max(probs, key=probs.get)

    # Transcribe the audio into text
    options = whisper.DecodingOptions()
    result = whisper.decode(model, mel, options)
    result_text = result.text

    # Validate detected language with langdetect (requires text)
    try:
        fallback_language = detect(result_text)
        if fallback_language != detected_language:
            detected_language = fallback_language
    except Exception as e:
        print(f"Language detection fallback failed: {e}")
        detected_language = 'en'  # Default to English

    return result_text, detected_language

# Enhanced text-to-speech function
def text_to_speech(text, language, file_path):
    # Fallback to English if the detected language is not supported by gTTS
    if language not in SUPPORTED_LANGUAGES:
        print(f"Language '{language}' not supported by gTTS, falling back to English.")
        language = 'en'

    try:
        audio_obj = gTTS(text=text, lang=language, slow=False)
        audio_obj.save(file_path)
    except Exception as e:
        print(f"Error in text-to-speech synthesis: {e}")
        return None
    return file_path

# Enhanced process_inputs function
def process_inputs(audio_path, image_path):
    try:
        # Process the audio file and detect the language
        speech_to_text_output, detected_language = transcribe(audio_path)
    except Exception as e:
        print(f"Error in transcribing audio: {e}")
        return "Error processing audio.", "Error detected.", None

    # Handle the image input
    if image_path:
        # Generate AI output based on the speech-to-text output and image
        try:
            chatgpt_output = img2txt(speech_to_text_output, image_path)
        except Exception as e:
            print(f"Error processing image: {e}")
            chatgpt_output = "Error processing image."
    else:
        chatgpt_output = "No image provided."

    # Use the detected language for text-to-speech synthesis
    try:
        processed_audio_path = text_to_speech(chatgpt_output, detected_language, "Temp3.mp3")
    except Exception as e:
        print(f"Error in text-to-speech synthesis: {e}")
        processed_audio_path = None

    return speech_to_text_output, chatgpt_output, processed_audio_path
```

```python
# Create the interface
iface = gr.Interface(
    fn=process_inputs,
    inputs=[
        gr.Audio(sources=["microphone"], type="filepath", label="Input Audio"),
        gr.Image(type="filepath", label="Input Image")
    ],
    outputs=[
        gr.Textbox(label="Speech to Text"),
        gr.Textbox(label="AI Output"),
        gr.Audio(label="Generated Audio Response")
    ],
    title="LLM Powered Voice Assistant for Multimodal Data",
    description="Upload an image and interact via voice input and audio response. The assistant responds in the language spoken in the audio."
)

# Launch the interface
iface.launch(debug=True)
```
